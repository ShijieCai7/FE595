```python
import numpy,matplotlib
from numpy import sin,cos,pi
from matplotlib import pyplot as plt
```

```python
x = numpy.linspace(-pi,pi,100)
ysin=sin(x)
ycos=cos(x)
```

```python
def Create_plot(a,b,c):
    plt.plot(a,b)
    plt.plot(a,c)
    plt.ylabel("y")
    plt.xlabel("x")
    plt.legend(('sin(x)','cos(x)'))
    plt.title('graph of sin(x) and cos(x) from -pi to pi')
    plt.show()
```

```python
Create_plot(x,ysin,ycos)
```
